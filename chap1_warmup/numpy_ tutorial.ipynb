24. 画图，y=x*x 其中 x = np.arange(0, 100, 0.1) （提示这里用到  matplotlib.pyplot 库）

```python
import matplotlib.pyplot as plt
x = np.arange(0, 100, 0.1)
y = x * x
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y, label='y = x²', color='tab:blue', linewidth=2)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y = x²', fontsize=12)
ax.set_title('Plot of y = x²', fontsize=14)
ax.grid(True, linestyle='--', alpha=0.7)
ax.legend()
plt.tight_layout()
plt.show()
```

25. 画图。画正弦函数和余弦函数， x = np.arange(0, 3 * np.pi, 0.1)(提示：这里用到 np.sin() np.cos() 函数和 matplotlib.pyplot 库)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(12, 7))  # 设置图形大小，使其清晰
plt.plot(x, y_sin, label='Sine(sin(x))', color='blue', linewidth=2)
plt.plot(x, y_cos, label='Cosine(cos(x))', color='red', linestyle='--', linewidth=2)

plt.xlabel('x (radians)', fontsize=12)
plt.ylabel('Amplitude', fontsize=12)
plt.title('Sine and Cosine Waves from 0 to 3π', fontsize=14)

pi_ticks = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi, 5*np.pi/2, 3*np.pi]
pi_labels = ['0', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$', r'$5\pi/2$', r'$3\pi$']
plt.xticks(pi_ticks, pi_labels, fontsize=10)

plt.legend()

plt.grid(True, linestyle='-', alpha=0.5)

plt.show()
```
